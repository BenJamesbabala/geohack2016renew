```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
sns.set_context('talk')



%matplotlib inline
```

```python
df = pd.read_csv('../data/trainingData.csv')
```

```python
wind = 0
solar = 1
geotherm = 2
rnw_palette = ['#5e93ca', "#fbc93d", "#fb4f4f"]
```

```python
df.head()
# b1 > thermal 
# b1_1 > population
```

```python
df_wind = df[df['nameindex'] == 0]
df_solar = df[df['nameindex'] == 1]
df_geotherm = df[df['nameindex'] == 2]
```

Distribution of mean solar irradiance at existing renewable site locations.

```python
fig, ax = plt.subplots(1, 3, figsize=(16,4))

ax[0].hist(df_wind['srad'], bins=30, range=(140,260), color=rnw_palette[0])
ax[0].set_ylabel('Number of Records')
ax[0].set_xlabel('Mean solar Irradiance')
ax[0].set_ylim(0,180)
ax[1].hist(df_solar['srad'], bins=30, range=(140,260), color=rnw_palette[1])
ax[1].set_xlabel('Mean solar Irradiance')
ax[1].set_ylim(0,180)
ax[2].hist(df_geotherm['srad'], bins=30, range=(140,260), color=rnw_palette[2])
ax[2].set_xlabel('Mean solar Irradiance')
ax[2].set_ylim(0,180)
plt.tight_layout()
```

```python
print("Median irradiance at wind farm locations:", df_wind['srad'].median())
print("Std dev of irradiance at wind farm locations:", df_wind['srad'].std())
print("Median irradiance at solar farm locations:", df_solar['srad'].median())
print("Std dev of irradiance at solar farm locations:", df_solar['srad'].std())
print("Median irradiance at geothermal plant locations:", df_geotherm['srad'].median())
print("Std dev of irradiance at geothermal plant locations:", df_geotherm['srad'].std())
```

Distribution of mean wind speeds at existing renewable site locations.

```python
fig, ax = plt.subplots(1, 3, figsize=(16,4))

ax[0].hist(df_wind['vs'], bins=30, range=(2,7), color=rnw_palette[0])
ax[0].set_ylabel('Number of Records')
ax[0].set_xlabel('Mean wind speed (m/s)')
ax[0].set_ylim(0,180)
ax[1].hist(df_solar['vs'], bins=30, range=(2,7), color=rnw_palette[1])
ax[1].set_xlabel('Mean wind speed (m/s)')
ax[1].set_ylim(0,180)
ax[2].hist(df_geotherm['vs'], bins=30, range=(2,7), color=rnw_palette[2])
ax[2].set_xlabel('Mean wind speed (m/s)')
ax[2].set_ylim(0,180)
plt.tight_layout()
```

```python
print("Median wind speed at wind farm locations:", df_wind['vs'].median())
print("Std dev of irradiance at wind farm locations:", df_wind['vs'].std())
print("Median wind speed at solar farm locations:", df_solar['vs'].median())
print("Std dev of irradiance at solar farm locations:", df_solar['vs'].std())
print("Median wind speed at geothermal plant locations:", df_geotherm['vs'].median())
print("Std dev of irradiance at geothermal plant locations:", df_geotherm['vs'].std())
```

Geothermal heat at existing renewable site locations.

```python
fig, ax = plt.subplots(1, 3, figsize=(16,4))

ax[0].hist(df_wind['b1'], bins=40, range=(0,10), color=rnw_palette[0])
ax[0].set_ylabel('Number of Records')
ax[0].set_xlabel('Geothermal Gradient (degrees C per km)')
#ax[0].set_ylim(0,180)
ax[1].hist(df_solar['b1'], bins=40, range=(0,10), color=rnw_palette[1])
ax[1].set_xlabel('Geothermal Gradient (degrees C per km)')
#ax[1].set_ylim(0,180)
ax[2].hist(df_geotherm['b1'], bins=40, range=(0,10), color=rnw_palette[2])
ax[2].set_xlabel('Geothermal Gradient (degrees C per km)')
#ax[2].set_ylim(0,180)
plt.tight_layout()
```

Population density at existing renewable plants.

```python
fig, ax = plt.subplots(1, 3, figsize=(16,4))

ax[0].hist(df_wind['b1_1'], bins=40, range=(0,100000), color=rnw_palette[0])
ax[0].set_ylabel('Number of Records')
ax[0].set_xlabel('Population Density')
ax[0].set_ylim(0,100)
ax[1].hist(df_solar['b1_1'], bins=40, range=(0,100000), color=rnw_palette[1])
ax[1].set_xlabel('Population Density')
ax[1].set_ylim(0,100)
ax[2].hist(df_geotherm['b1_1'], bins=40, range=(0,100000), color=rnw_palette[2])
ax[2].set_xlabel('Population Density')
ax[2].set_ylim(0,100)
plt.tight_layout()
```
