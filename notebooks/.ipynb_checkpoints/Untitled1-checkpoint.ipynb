```python
import pandas as pd
from osgeo import ogr, osr
```

```python
shp = '/Users/SamHooper/Grad_School/geohackweek/renewables/PowerPlants_US_EIA/PowerPlants_US_201603.shp'
```

```python
def fc_to_df(fc, out_txt=None):
    
    
    fields = [f.name for f in arcpy.ListFields(fc) if f.name != 'Shape']

    vals = []
    cursor = arcpy.SearchCursor(fc)
    for row in arcpy.SearchCursor(fc):
        these_vals = [row.getValue(f) for f in fields]
        vals.append(these_vals)
    del row, cursor
    
    df = pd.DataFrame(vals, columns=fields)
    if out_txt:
        df.to_csv(out_txt, index=False)
    return df
```

```python
def shp_to_df(shp, out_txt=None):
    ds = ogr.Open(shp)
    lyr = ds.GetLayer()
    lyr_def = lyr.GetLayerDefn()
    fields = [lyr_def.GetFieldDefn(i).GetName() for i in range(lyr_def.GetFieldCount())]

    vals = []
    #feat = lyr.GetNextFeature()
    #print feat.
    for feat in lyr:
        these_vals = [feat.GetField(f) for f in fields]
        vals.append(these_vals)
        #feat = lyr.GetNextFeature()
    del feat
    
    df = pd.DataFrame(vals, columns=fields)
    if out_txt:
        df.to_csv(out_txt, index=False)
    return df
```

```python
df = shp_to_df(shp)
```
